<h2>Train models</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
X = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']
```

```python
df.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101, stratify=True)
```

<h3>Logistic Regression</h3>

```python
from sklearn.linear_model import LogisticRegression
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(X_train, y_train)
```

```python
predictions = logistic_regression_model.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

```python
sns.heatmap(conf_matrix, annot=True)
```

<h3>KNeighbors</h3>

```python
predictions = knn_model.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

```python
sns.heatmap(conf_matrix, annot=True)
```

<h3>Neural Nets</h3>

```python
X = df.drop('Bankrupt?', axis=1).values
y = df['Bankrupt?'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model_neuralnets = Sequential()

model_neuralnets.add(Dense(20, activation='relu'))
model_neuralnets.add(Dense(20, activation='relu'))
model_neuralnets.add(Dense(20, activation='relu'))
model_neuralnets.add(Dense(20, activation='relu'))

model_neuralnets.add(Dense(1))

model_neuralnets.compile(optimizer='adam', loss='mse')
```

```python
model_neuralnets.fit(
    x=X_train,
    y=y_train,
    epochs=400
)
```

```python
loss_df = pd.DataFrame(model_neuralnets.history.history)
loss_df.plot()
```

```python
model_neuralnets.evaluate(X_test, y_test, verbose=0)
```

```python
model_neuralnets.evaluate(X_train, y_train, verbose=0)
```

```python
predictions = model_neuralnets.predict(X_test)
```

```python
predictions
```

```python
predictions = (model_neuralnets.predict(X_test) > 0.5).astype(int)
```

```python
print(classification_report(y_test, predictions))
```

```python
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

```python
sns.heatmap(conf_matrix, annot=True)
```
