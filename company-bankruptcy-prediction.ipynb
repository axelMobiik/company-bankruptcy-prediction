Train

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic_regression_model = LogisticRegression()
```

```python
logistic_regression_model.fit(X_train, y_train)
```

```python
predictions = logistic_regression_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

KNeighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
predictions = decision_tree.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor(n_estimators=200)
```

```python
random_forest.fit(X_train, y_train)
```

```python
predictions = random_forest.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

Neural Nets

```python
X = df.drop('Bankrupt?', axis=1).values
y = df['Bankrupt?'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model2 = Sequential()

model2.add(Dense(20, activation='relu'))
model2.add(Dense(20, activation='relu'))
model2.add(Dense(20, activation='relu'))
model2.add(Dense(20, activation='relu'))

model2.add(Dense(1))

model2.compile(optimizer='adam', loss='mse')
```

```python
model2.fit(
    x=X_train,
    y=y_train,
    epochs=400
)
```

```python
loss_df = pd.DataFrame(model2.history.history)
loss_df.plot()
```

```python
model2.evaluate(X_test, y_test, verbose=0)
```

```python
model2.evaluate(X_train, y_train, verbose=0)
```

```python
test_predictions = model2.predict(X_test)
```

```python
test_predictions
```

```python
test_predictions = (model2.predict(X_test) > 0.5).astype(int)
```

```python
print(classification_report(y_test, test_predictions))
```

```python
print(confusion_matrix(y_test, test_predictions))
```
