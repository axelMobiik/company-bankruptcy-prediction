```python
def columns_unique_value(df):
    new_df = 0
    
    for column in df.columns:
        if len(df[column].unique()) <= 1:
            new_df = df.drop(column, axis=1)
    
    return new_df

df = columns_unique_value(df)
df.columns
```

```python
bankrupt_true = df[df['Bankrupt?']==1]
bankrupt_false = df[df['Bankrupt?']==0]
```

```python
bankrupt_true.shape
```

```python
bankrupt_false.shape
```

```python
bankrupt_true = bankrupt_true.sample(n=6599, replace=True, random_state=101)
```

```python
bankrupt_true.shape
```

```python
df = pd.concat([bankrupt_false, bankrupt_true])
```

```python
df.shape
```

```python
df
```

<h1>Train models</h1>

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from skopt import BayesSearchCV
from skopt.space import Categorical, Real

X = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']
```

```python
X.head()
```

```python
y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101, stratify=y)
```

<h2>Logistic Regression</h2>

```python
from sklearn.linear_model import LogisticRegression
param_bayes = {
    'solver': Categorical(['newton-cg', 'sag', 'liblinear']),
    'penalty': Categorical(['l2']),
    'C': Real(0.01, 100, "log-uniform")
}

logistic_regression_model = LogisticRegression()
bayes_search = BayesSearchCV(logistic_regression_model, param_bayes, scoring='f1', random_state=2)
bayes_search.fit(X_train, y_train)

print('Best hyperparameters: ', bayes_search.best_params_)
print('Best score: ', bayes_search.best_score_)
```

```python
logistic_regression_model = LogisticRegression(penalty='l2', C=100, solver='sag')
logistic_regression_model.fit(X_train, y_train)
```

```python
predictions = logistic_regression_model.predict(X_test)
class_report_logistic_regression = classification_report(y_test, predictions)
conf_matrix_logistic_regression = confusion_matrix(y_test, predictions)
print(class_report_logistic_regression)
print('\n')
print(conf_matrix_logistic_regression)
```

<h2>KNeighbors</h2>

```python
from sklearn.neighbors import KNeighborsClassifier

param_grid = {
    'n_neighbors': (1,10, 1),
    'leaf_size': (20,40,1),
    'p': (1,2),
    'weights': ('uniform', 'distance'),
    'metric': ('minkowski', 'chebyshev')
}

knn_model = KNeighborsClassifier()
grid_search = GridSearchCV(knn_model, param_grid, cv=10, scoring='accuracy')
grid_search.fit(X_train, y_train)

print('Best hyperparameters: ', grid_search.best_params_)
print('Best score: ', grid_search.best_score_)
```

```python
knn_model = KNeighborsClassifier(leaf_size=20, metric='minkowski', n_neighbors=1, p=1, weights='uniform')
knn_model.fit(X_train, y_train)
```

```python
predictions = knn_model.predict(X_test)
class_report_knn = classification_report(y_test, predictions)
conf_matrix_knn = confusion_matrix(y_test, predictions)
print(class_report_knn)
print('\n')
print(conf_matrix_knn)
```

<h2>Decision Tree Classifier</h2>

```python
from sklearn.tree import DecisionTreeClassifier

param_grid_decision_tree = {
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'criterion': ["gini", "entropy"]
}

decision_tree_model = DecisionTreeClassifier()
grid_search = GridSearchCV(decision_tree_model, param_grid_decision_tree, cv=10, scoring='accuracy')
grid_search.fit(X_train, y_train)

print('Best hyperparameters: ', grid_search.best_params_)
print('Best score: ', grid_search.best_score_)
```

```python
decision_tree_model = DecisionTreeClassifier(criterion='entropy', max_depth=20, min_samples_leaf=5)
decision_tree_model.fit(X_train, y_train)
```

```python
predictions = decision_tree_model.predict(X_test)
class_report_decision_tree = classification_report(y_test, predictions)
conf_matrix_decision_tree = confusion_matrix(y_test, predictions)
print(class_report_decision_tree)
print('\n')
print(conf_matrix_decision_tree)
```

<h2>Random Forest Classifier</h2>

<h2>Neural Nets - Keras</h2>

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
X_train_scaler.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

neuralnets_model = Sequential()

neuralnets_model.add(Dense(20, activation='relu'))
neuralnets_model.add(Dense(20, activation='relu'))
neuralnets_model.add(Dense(20, activation='relu'))
neuralnets_model.add(Dense(20, activation='relu'))

neuralnets_model.add(Dense(1))

neuralnets_model.compile(optimizer='adam', loss='mse')
```

```python
neuralnets_model.fit(
    x=X_train_scaler,
    y=y_train,
    epochs=200
)
```

```python
loss_df = pd.DataFrame(neuralnets_model.history.history)
loss_df.plot()
```

```python
neuralnets_model.evaluate(X_test_scaler, y_test, verbose=0)
```

```python
neuralnets_model.evaluate(X_train_scaler, y_train, verbose=0)
```

```python
predictions = neuralnets_model.predict(X_test_scaler)
predictions
```

```python
predictions = (neuralnets_model.predict(X_test_scaler) > 0.5).astype(int)
class_report_neuralnets = classification_report(y_test, predictions)
conf_matrix_neuralnets = confusion_matrix(y_test, predictions)
print(class_report_neuralnets)
print('\n')
print(conf_matrix_neuralnets)
```
